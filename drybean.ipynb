```python
#数据预处理
X = df.iloc[:, :16]
Y = df.iloc[:, 16:]
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split

valSize = round(len(X) * 0.1)

X_train, X_test, Y_train, Y_test = train_test_split(X[:-valSize], Y[:-valSize], test_size=0.4, random_state=1)

X_val = X[-valSize:]
Y_val = Y[-valSize:]

print(Y_train.shape)
print(Y_test.shape)
print(Y_val.shape)
```

```python
from sklearn import preprocessing

stdProcess = preprocessing.StandardScaler()
stdProcess.fit(X_train)
```

```python
process_X_train = stdProcess.transform(X_train, copy=True)
process_X_test = stdProcess.transform(X_test, copy=True)
process_X_val = stdProcess.transform(X_val, copy=True)
```

```python
from sklearn.linear_model import LogisticRegression

maxAccuracy = 0.0
highModel = None
highPred = None
for solver in ['newton-cg', 'lbfgs', 'sag', 'saga']:
    model = LogisticRegression(penalty='l2', solver=solver, multi_class='multinomial', max_iter=2000)
    model.fit(process_X_train, Y_train.squeeze())

    predictRes = model.predict(process_X_test)

    groundTruth = Y_test.squeeze()
    groundTruth = groundTruth.to_numpy()

    res = groundTruth == predictRes
    accuracy = model.score(process_X_test, groundTruth)
    print("accuracy %.10f" % (accuracy))
    if accuracy > maxAccuracy:
        print('solver:{}'.format(solver))
        highModel = model
        highPred = predictRes
        maxAccuracy = accuracy
```

```python
#绘制混淆矩阵
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt

y_pred = highPred  # ['2','2','3','1','4'] # 类似的格式
y_true = groundTruth  # ['0','1','2','3','4'] # 类似的格式

C = confusion_matrix(y_true, y_pred, labels=['SEKER', 'BARBUNYA', 'BOMBAY', 'CALI', 'HOROZ', 'SIRA',
                                             'DERMASON'])  # 可将'1'等替换成自己的类别，如'cat'。

plt.matshow(C, cmap=plt.cm.Reds)  # 根据最下面的图按自己需求更改颜色
for i in range(len(C)):
    for j in range(len(C)):
        plt.annotate(C[j, i], xy=(i, j), horizontalalignment='center', verticalalignment='center')

plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.show()
```

```python
y_val_pred = highModel.predict(process_X_val)

y_val_groundtruth = Y_val.squeeze()
y_val_groundtruth = y_val_groundtruth.to_numpy()

val_res = y_val_groundtruth == y_val_pred
accuracy = highModel.score(process_X_val, y_val_groundtruth)
```

```python
print("val accuracy %.7f" % (accuracy))
```
